```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

column = ["Age", "Op_Year", "axil_nodes_det", "Surv_status"]
hpd = pd.read_csv('haberman.csv', names=column)

hpd.head(10)
```

```python
hpd.shape
```

```python
hpd.info()
```

```python
hpd.describe()
```

```python
hpd.groupby("Surv_status").size()
```

```python
sns.set_style("whitegrid")
sns.pairplot(data=hpd, hue="Surv_status" ,vars = ["Age", "Op_Year", "axil_nodes_det"],height=4)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hpd, hue="Surv_status", height=3).map(plt.scatter, "Age", "axil_nodes_det").add_legend();
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hpd, hue="Surv_status", height=3).map(plt.scatter, "Op_Year", "axil_nodes_det").add_legend();
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(hpd, hue="Surv_status", height=3).map(plt.scatter, "axil_nodes_det", "Surv_status").add_legend();
plt.show()
```

```python
sns.FacetGrid(hpd, hue="Surv_status", height=5) \
   .map(sns.distplot, "Age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(hpd, hue="Surv_status", height=5) \
   .map(sns.distplot, "Op_Year") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(hpd, hue="Surv_status", height=5) \
   .map(sns.distplot, "axil_nodes_det") \
   .add_legend();
plt.show();
```

```python
sns.boxplot(x='axil_nodes_det',y='Surv_status', data=hpd)
plt.show()
```
